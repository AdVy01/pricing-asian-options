# Example usage of the pricing methods

Perform the necessary imports

```python
from core.laplace_transform import calculate_asian_call_price
from core.monte_carlo_control_variate import AsianOptionMC_MC
from core.vecer_dimension_reduction import vecer_fixed_asian_call
from core.mpdata import calculate_mpdata_price
```

Define the option's parameters

```python
S0 = 100
K = 100
r = 0.05
T = 1.0
sigma = 0.20
```

## Monte Carlo With Control Variate

```python
# Monte Carlo Parameters
simulations = 400000
time_steps = 500
div = 0.0

# Initialize the Monte Carlo class
mc_sim = AsianOptionMC_MC(option_type='call', S0=S0, strike=K, T=T, M=time_steps, r=r, div=div, sigma=sigma, simulations=simulations)

# Get the price WITHOUT control variate
p, lower_bound, upper_bound = mc_sim.value

# Get the price WITH control variate
p_cv, lower_bound_cv, upper_bound_cv = mc_sim.value_with_control_variate

print(f"Monte Carlo WITHOUT control variate: {p} with the 95% confidence interval: ({lower_bound}, {upper_bound})")
print(f"Monte Carlo WITH control variate: {p_cv} with the 95% confidence interval: ({lower_bound_cv}, {upper_bound_cv})")
```

## Večeř dimension reduction

```python
# Vecer parameters
M = 1000
N = 500

p_vecer = vecer_fixed_asian_call(S0=S0, K=K, T=T, r=r, sigma=sigma, M=M, N=N)

print(f"Večeř dimension reductin price: {p_vecer}.")
```

## Laplace Transform

```python
# Laplace parameters
t0 = 0
t = 0
A = S0
S = S0

p_laplace = calculate_asian_call_price(r=r, sigma=sigma, T=T, t0=t0, t=t, A=A, S=S, K=K)

print(f"Laplace Transform price: {p_laplace}")
```

## MPDATA

```python
# MPDATA parameters
S_max = 300
S_min = 20
nt = 1800
nx = 80
ny = 80
scale = 4

p_mpdata = calculate_mpdata_price(T=T, K=K, r=r, sgma=sigma, S_max=S_max, S_min=S_min, nt=nt, nx=nx, ny=ny, scale=scale, S0=S0)

print(f"MPDATA price: {p_mpdata}")
```
